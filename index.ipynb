## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A Resilient Distributed Dataset(RDD) is a collection of multiple servers that share data from databases beetween all these servers.
Each server contains data about a proper database plus a backup of data of other databases.
Fault tollerance refers to the safe-method used from the RDD, basically if 1 or more servers are not working for some reasons this system is still able to access all the data of the database thanks to the backups.
